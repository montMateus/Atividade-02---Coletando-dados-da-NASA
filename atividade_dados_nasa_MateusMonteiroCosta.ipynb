```python
import requests

def receber_dados(data_inicial: str, data_final: str):
    api_key = "DEMO_KEY"
    url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={data_inicial}&end_date={data_final}&api_key={api_key}"

    resposta = requests.get(url)

    if resposta.status_code == 200:
        resposta_json = resposta.json()
        resposta_json = resposta_json["near_earth_objects"]
        meteoros = processar_dados(resposta_json)
        return mostrar_dados(meteoros)

    else:
        return {"Erro": "Não foi possível obter os dados"}
 
def processar_dados(resposta: dict):
    lista_meteoros = []

    for data, objetos in resposta.items():
        for obj in objetos:
            dados_processados = {
                'Nome do asteroide': obj.get('name'),
                'Diâmetro máximo (m)': obj['estimated_diameter']['meters']['estimated_diameter_max'],
                'Velocidade relativa (km/h)': obj['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'],
                'Distância mínima da Terra (km)': obj['close_approach_data'][0]['miss_distance']['kilometers'],
                'Potencialmente perigoso': "Sim" if obj['is_potentially_hazardous_asteroid'] else "Não"
            }
            lista_meteoros.append(dados_processados)

    return lista_meteoros

def mostrar_dados(lista_meteoros: list):
    i = 1
    for meteoros in lista_meteoros:
            print('')
            print(f'{i}º - {meteoros}')
            print('')
            print('-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
            i+=1

data_inicial = input('Digite a data inicial a ser procurada, no formato ANO-MES-DIA: ')
data_final = input('Digite a data final a ser procurada, no formato ANO-MES-DIA: ')

resultado = receber_dados(data_inicial=data_inicial, data_final=data_final)
print(resultado)
```
